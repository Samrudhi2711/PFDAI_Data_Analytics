<b>14. Write a Python program to create a stacked bar plot with error bars.

Means (women) = (25, 32, 30, 35, 29) 
Men Standard deviation = (4, 3, 4, 1, 5) 
Women Standard deviation = (3, 5, 2, 3, 3)</b>

```python
men_means = np.array([22, 30, 35, 35, 26])
women_means = np.array([25, 32, 30, 35, 29])
men_std = np.array([4, 3, 4, 1, 5])
women_std = np.array([3, 5, 2, 3, 3])

# X positions for the bars (e.g., categories 1 to 5)
ind = np.arange(len(men_means))  
width = 0.5                      # Bar width

# 2. Create the Stacked Bar Plot
plt.figure(figsize=(10, 7))

# Plot 1: Men (the base of the stack)
p1 = plt.bar(
    ind, 
    men_means, 
    width, 
    yerr=men_std, 
    capsize=5, 
    label='Men',
    color='skyblue',
    edgecolor='black'
)

# Plot 2: Women (stacked on top of men)
# The 'bottom=men_means' argument is key for stacking
p2 = plt.bar(
    ind, 
    women_means, 
    width, 
    bottom=men_means, # <-- Stacks women's bar on top of men's bar
    yerr=women_std, 
    capsize=5, 
    label='Women',
    color='lightcoral',
    edgecolor='black'
)

# 3. Customizations
plt.xlabel('Group/Category', fontsize=12)
plt.ylabel('Total Score', fontsize=12)
plt.title('Stacked Bar Plot of Men and Women Means with Error Bars (Q14)', fontsize=14)

# Set X ticks to be centered beneath the bars (index + label)
plt.xticks(ind, ['G1', 'G2', 'G3', 'G4', 'G5']) 

# Add the legend and grid
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

<b>15. Write a Python program to create a horizontal bar chart with differently 
ordered colors. 
Note: Use bottom to stack the women?s bars on top of the men?s bars. 
Sample Data Set: 
languages = [['Language','Science','Math'], 
['Science','Math','Language'], 
['Math','Language','Science']] 
numbers = [{'Language':75, 'Science':88, 'Math':96}, 
{'Language':71, 'Science':95, 'Math':92}, 
{'Language':75, 'Science':90, 'Math':89}]</b>

```python
import matplotlib.pyplot as plt
import numpy as np

languages = [
    ['Language', 'Science', 'Math'],
    ['Science', 'Math', 'Language'],
    ['Math', 'Language', 'Science']
]
numbers = [
    {'Language': 75, 'Science': 88, 'Math': 96},
    {'Language': 71, 'Science': 95, 'Math': 92},
    {'Language': 75, 'Science': 90, 'Math': 89}
]

students = ['Student-1', 'Student-2', 'Student-3']

color_map = {
    'Language': 'red',
    'Math': 'blue',
    'Science': 'green'
}

fig, ax = plt.subplots(figsize=(10, 5))
y_pos = np.arange(len(students))
bar_height = 0.5

for i, student in enumerate(students):
    left_position = 0
    current_order = languages[i]
    
    for subject in current_order:
        score = numbers[i][subject]
        color = color_map[subject]
        
        ax.barh(
            y_pos[i],
            score,
            height=bar_height,
            left=left_position,
            color=color,
            label=subject if i == len(students) - 1 else ""
        )
        
        left_position += score

ax.set_yticks(y_pos)
ax.set_yticklabels(students)

ax.set_xlim(0)

handles, labels = ax.get_legend_handles_labels()

legend_order = ['Language', 'Math', 'Science']
ordered_handles = [handles[labels.index(key)] for key in legend_order]
ax.legend(ordered_handles, legend_order, loc='upper right', bbox_to_anchor=(0.95, 0.95), frameon=True)

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

plt.tight_layout()
plt.show()
```

<b>16. Write a Python program to create stack bar plot and add label to each 
section. 
Sample data: 
people = ('G1','G2','G3','G4','G5','G6','G7','G8') 
segments = 4 
multi-dimensional data 
data = [[ 3.40022085, 7.70632498, 6.4097905, 10.51648577, 7.5330039, 
7.1123587, 12.77792868, 3.44773477], 
[ 11.24811149, 5.03778215, 6.65808464, 12.32220677, 7.45964195, 
6.79685302, 7.24578743, 3.69371847], 
[ 3.94253354, 4.74763549, 11.73529246, 4.6465543, 12.9952182, 
4.63832778, 11.16849999, 8.56883433], 
[ 4.24409799, 12.71746612, 11.3772169, 9.00514257, 10.47084185, 
10.97567589, 3.98287652, 8.80552122]]</b>

```python
import numpy as np
import matplotlib.pyplot as plt

people = ('G1', 'G2', 'G3', 'G4', 'G5', 'G6', 'G7', 'G8')
segments = 4

data = np.array([
    [3.40022085, 7.70632498, 6.4097905, 10.51648577, 7.5330039, 7.123587, 12.7792868, 3.44773477],
    [11.2481149, 5.03778215, 6.65808464, 12.3220677, 7.45964195, 6.79685302, 7.24578743, 3.69371847],
    [3.94253354, 4.74763549, 11.73529246, 4.6465543, 12.9952182, 4.63832778, 11.16849999, 8.56883433],
    [4.24409799, 12.71746612, 11.3772169, 9.00514257, 10.47084185, 10.97567589, 3.98287652, 8.80552122]
]).T

totals = data.sum(axis=1)

data_perc = data * 100 / totals[:, np.newaxis]

fig, ax = plt.subplots(figsize=(10, 6))
lefts_raw = np.zeros(len(people))

colors = ['#d62728', '#2ca02c', '#f7f7f7', '#9467bd']

for i in range(segments):
    ax.barh(
        people,
        data[:, i],
        left=lefts_raw,
        color=colors[i],
        edgecolor='black',
        linewidth=0.5
    )

    x_centers_raw = lefts_raw + data[:, i] / 2

    for j, (x, y) in enumerate(zip(x_centers_raw, people)):
        percentage = data_perc[j, i]
        if percentage > 4:
            ax.text(
                x, y,
                f'{percentage:.0f}%',
                ha='center',
                va='center',
                color='black',
                fontsize=9
            )

    lefts_raw += data[:, i]

ax.set_xlabel('Scores')
ax.invert_yaxis()
ax.set_xlim(0)

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

plt.title('Stack Bar Plot with Segment Labels')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


# Number of bars (stacks)
num_bars = 10
x = np.arange(num_bars) # x-coordinates for the bars

# Base data (e.g., 'mens bars') - all have a height of 1.0
mens_heights = np.array([1.0] * num_bars)

# Stacking data (e.g., 'womens bars') - all have a height of 2.0
womens_heights = np.array([2.0] * num_bars)

# --- Hatch Patterns ---
# A list of black and white hatch patterns to match the screenshot's textures.
# Matplotlib offers a variety of patterns like '/', '\', '|', '-', '+', 'x', 'o', 'O', '.', '*'
hatches = ['/', '\\', '|', '-', '+', 'x', 'o', 'O', '.', '*']

mens_hatches = ['/', '\\', '|', '-', '+', 'x', 'o', 'O', '.', '*']


mens_hatches_visual = ['/', '\\', '|', '-', '+', 'x', 'o', 'O', '.', '*']
womens_hatches_visual = ['//', '\\\\', '||', '--', '++', 'xx', 'oo', 'OO', '..', '**']

hatches_for_image_match = [
    '/', '\\', '|', ':', '||', 'x', 'o', '.', 'O', '*'
]
# We'll cycle through a broader list of available patterns for better variance.
hatches_for_image_match = [
    '/', '\\', 'x', '+', '-', '.', '*', 'o', 'O', '||'
]

# --- Plotting ---
fig, ax = plt.subplots(figsize=(8, 6))


mens_rects = []
for i in range(num_bars):
    rect = ax.bar(
        x[i], # x-position
        mens_heights[i], # height
        width=0.8, # width of the bar
        color='white', # Set color to white for black and white texture
        edgecolor='black', # Black outline
        hatch=hatches_for_image_match[i], # Apply the specific hatch pattern
        zorder=2 # Ensure bars are on top
    )
    mens_rects.append(rect)

# 2. Plot the 'womens bars' stacked on top of the 'mens bars'
# Use the 'bottom' argument to stack 'womens bars' on top of 'mens bars'.
womens_rects = []
for i in range(num_bars):
    rect = ax.bar(
        x[i], # x-position
        womens_heights[i], # height
        width=0.8, # width of the bar
        bottom=mens_heights[i], # **Crucial for stacking**
        color='white',
        edgecolor='black',
        # Applying the SAME hatch pattern to the top part to achieve the unified look of the screenshot
        hatch=hatches_for_image_match[i],
        zorder=2
    )
    womens_rects.append(rect)


ax.set_xticks(x)
ax.set_xticklabels(x) # Label the bars from 0 to 9

ax.set_ylim(0, 3.0)
ax.set_yticks(np.arange(0, 3.5, 0.5)) # Ticks at 0.0, 0.5, 1.0, 1.5, 2.0, 2.5, 3.0

ax.set_xlim(-0.5, 10)
ax.set_xticks(np.arange(0, 11, 2))
ax.set_xticklabels(np.arange(0, 11, 2))

# Remove top and right spines
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

plt.tight_layout()

plt.show()
```
