<b>14. Write a Python program to display the grid and draw line charts of the
closing value of Alphabet Inc. between October 3, 2016 to October 7, 2016.
Customized the grid lines with rendering with a larger grid (major grid) and a
smaller grid (minor grid).Turn on the grid but turn off ticks.
The code snippet gives the output shown in the following screenshot:</b>

```python
import matplotlib.dates as mdates
dates = ['2016-10-03', '2016-10-04', '2016-10-05', '2016-10-06', '2016-10-07']
values = [772.75, 776.40, 776.45, 776.80, 775.10]
date_objects = pd.to_datetime(dates)

fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(
    date_objects,
    values,
    color='red',
    marker='o',         
    linestyle='-',
    linewidth=1.5,
    markersize=6        
)

ax.set_title('Closing stock value of Alphabet Inc.', fontsize=14)
ax.set_xlabel('Date', fontsize=12)
ax.set_ylabel('Closing Value', fontsize=12)


ax.minorticks_on()

ax.grid(which='major', axis='both', linestyle='-', linewidth=0.5, color='gray', alpha=0.5)

ax.grid(which='minor', axis='both', linestyle='-', linewidth=0.2, color='gray', alpha=0.3)

ax.tick_params(which='both', length=0)
ax.set_ylim(772.5, 777.0)
ax.set_yticks(np.arange(772.5, 777.5, 0.5))

ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
fig.autofmt_xdate(rotation=0, ha='center')
plt.tight_layout()
plt.show()
```

<b>15. Write a Python program to create multiple plots.
The code snippet gives the output shown in the following screenshot:</b>

```python
x = np.linspace(0, 2 * np.pi, 50)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x / 4)
y4 = x

fig = plt.figure(figsize=(12, 7))

ax1 = plt.subplot(2, 3, (1, 3)) 
ax1.plot(x, y4, color='blue', label='Top Plot: Linear Trend')
ax1.set_title('Large Top Plot (2x3 Grid, spanning all 3 columns)')
ax1.set_ylabel('Y4 Data')
ax1.grid(True, linestyle=':', alpha=0.6)


ax2 = plt.subplot(2, 3, 4)
ax2.plot(x, y1, color='red', label='Subplot 1: Sine')
ax2.set_title('Subplot 1 (2,3,4)')
ax2.set_xlabel('X')
ax2.grid(True, linestyle=':', alpha=0.6)

ax3 = plt.subplot(2, 3, 5)
ax3.plot(x, y2, color='green', label='Subplot 2: Cosine')
ax3.set_title('Subplot 2 (2,3,5)')
ax3.set_xlabel('X')
ax3.grid(True, linestyle=':', alpha=0.6)

ax4 = plt.subplot(2, 3, 6)
ax4.plot(x, y3, color='orange', label='Subplot 3: Tan')
ax4.set_title('Subplot 3 (2,3,6)')
ax4.set_xlabel('X')
ax4.grid(True, linestyle=':', alpha=0.6)

plt.tight_layout()
plt.show()

```
